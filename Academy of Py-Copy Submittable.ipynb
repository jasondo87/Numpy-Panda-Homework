```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#reference file and make it a variable
student_file = "./raw_data/students_complete.csv"
school_file = "./raw_data/schools_complete.csv"
```

```python
#import the data in Panda Dataframe
student_df = pd.read_csv(student_file)
school_df = pd.read_csv(school_file)
```

```python
#check columns and indexes by reading the file
student_df.head()
```

```python
#check columns and indexes by reading the file
school_df.head()
#change column name "name" to "school
school_df = school_df.rename(columns = {"name": "school",
                                       "type": "School Type",
                                       "budget": "Total School Budget"})
school_df.head()
```

```python
#What are the total schools?
total_schools = len(student_df["school"].unique())
#What are total students?
total_students = len(student_df["name"])
#What are the total budgets?
total_budget = school_df["Total School Budget"].sum()
#What is the average math score?
math_average = student_df["math_score"].mean()
#What is the average reading score?
read_average = student_df["reading_score"].mean()
#What is the % passing math
math_pass_count = len(student_df.loc[student_df["math_score"] >= 70])
math_score_count = len(student_df["math_score"])
math_pass = (math_pass_count/math_score_count)* 100
#What is the % passing reading
read_pass_count = len(student_df.loc[student_df["reading_score"] >= 70])
read_score_count = len(student_df["reading_score"])
read_pass = (read_pass_count/read_score_count)* 100
#What is the overall passing rate, average of the two?
average_pass_score = (math_average + read_average)/2
```

```python
#input the summaries into a dataframe
snapshot_summary_df = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                "Average Math Score":[math_average],
                                "Average Reading Score": [read_average],
                                "% Passing Math": [math_pass],
                                "% Passing Reading": [read_pass],
                                "Pass Score Averages": [average_pass_score]})
#sort out all the columns
snapshot_summary_df = snapshot_summary_df[["Total Schools",
                                    "Total Students",
                                    "Total Budget",
                                    "Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading",
                                    "Pass Score Averages"]]

#map formatting
snapshot_summary_df["Total Students"] = snapshot_summary_df["Total Students"].map("{:,}".format)
snapshot_summary_df["Total Budget"] = snapshot_summary_df["Total Budget"].map("${0:,.0f}".format)
snapshot_summary_df["% Passing Math"] = snapshot_summary_df["% Passing Math"].map("{0:,.2f}%".format)
snapshot_summary_df["% Passing Reading"] = snapshot_summary_df["% Passing Reading"].map("{0:,.2f}%".format)

snapshot_summary_df
```

```python
#Dataframe for math pass
math_pass_df = student_df.loc[student_df["math_score"]>=70]
#Dataframe for read pass
read_pass_df = student_df.loc[student_df["reading_score"]>=70]


```

```python
#create groupby for math > 70
group_math_pass = math_pass_df.groupby(["school"])
pass_math_count = group_math_pass["math_score"].count()
#create groupby for reading > 70

group_read_pass = read_pass_df.groupby(["school"])
pass_read_count = group_read_pass["reading_score"].count()

```

```python
#groupby by school to get the pivot calculations
grouped_schools_df = student_df.groupby(["school"])
schools_students = grouped_schools_df["school"].count()
schools_math_average = grouped_schools_df["math_score"].mean()
schools_read_average = grouped_schools_df["reading_score"].mean()
average_math_pass = (pass_math_count/schools_students)*100
average_read_pass = (pass_read_count/schools_students)*100
overall_pass = (average_math_pass + average_read_pass)/(2)


#turn the groupbys to a datafram so we can do calculations
school_summary = pd.DataFrame({"Total Students": schools_students, 
                               "Average Math Score": schools_math_average,
                              "Average Reading Score": schools_read_average,
                               "% Passing Math": average_math_pass, 
                               "% Passing Reading": average_read_pass,
                              "% Overall Passing Rate": overall_pass})

school_summary.reset_index(inplace=True)
school_summary



```

```python
#merge the two databases
summary_merged = pd.merge(school_summary, school_df, on="school")
summary_merged



```

```python
#calculate per student budget
summary_merged["Per Student Budget"]= summary_merged["Total School Budget"]/summary_merged["Total Students"]
#clean data
#del summary_merged["size"]
summary_merged_rename = summary_merged.rename(columns={"school":"School"})
summary_merged_rename

#reorganize the table
summary_merged_rename = summary_merged_rename[["School", "School Type", "Total Students", "Total School Budget", 
                                 "Per Student Budget", "Average Math Score", "Average Reading Score",
                                "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]

summary_merged_rename
```

```python
#map formatting
summary_merged_rename["Total Students"] = summary_merged_rename["Total Students"].map("{:,}".format)
summary_merged_rename["Total School Budget"] = summary_merged_rename["Total School Budget"].map("${0:,.0f}".format)
summary_merged_rename["Per Student Budget"] = summary_merged_rename["Per Student Budget"].map("${0:,.0f}".format)
summary_merged_rename["% Passing Reading"] = summary_merged_rename["% Passing Reading"].map("{0:,.2f}%".format)
summary_merged_rename["% Passing Math"] = summary_merged_rename["% Passing Math"].map("{0:,.2f}%".format)
summary_merged_rename["% Overall Passing Rate"] = summary_merged_rename["% Overall Passing Rate"].map("{0:,.2f}%".format)
```
